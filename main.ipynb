```python
# Imports

import pandas as pd
from matplotlib import *
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_spss('dennis spss data.sav')
```

```python
def compute_frequencies_percentages(column):
    frequency_counts = column.value_counts()
    percentages = column.value_counts(normalize=True) * 100
    return pd.DataFrame({'Frequency': frequency_counts, 'Percentage': percentages})

# Analyze categorical columns
categorical_columns = ['VAR00001', 'VAR00002', 'VAR00003', 'VAR00004', 'VAR00015', 'VAR00016', 'VAR00017', 'VAR00021', 'VAR00022', 'VAR00019']
for col in categorical_columns:
    print(f'Analysis for {col}:')
    print(compute_frequencies_percentages(df[col]))
    print()

# Analyze numeric columns
numeric_columns = ['revenuealloc_comp', 'revenuexpd_comp', 'disasteroutcome_comp']
for col in numeric_columns:
    print(f'Analysis for {col}:')
    mean = df[col].mean()
    std_dev = df[col].std()
    print(f'Mean: {mean}')
    print(f'Standard Deviation: {std_dev}')
    print()

# If you want to calculate means and std dev for other numeric columns
df.describe(include=[np.number])
```

```python
# Function to plot frequency and percentage for a categorical column
def plot_categorical_data(column, title):
    freq_data = compute_frequencies_percentages(df[column])
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))

    # Frequency Plot
    sns.barplot(x=freq_data.index, y=freq_data['Frequency'], ax=ax[0], palette='viridis')
    ax[0].set_title(f'Frequency of {title}')
    ax[0].set_xlabel(title)
    ax[0].set_ylabel('Frequency')
    ax[0].tick_params(axis='x', rotation=45)

    # Percentage Plot
    sns.barplot(x=freq_data.index, y=freq_data['Percentage'], ax=ax[1], palette='viridis')
    ax[1].set_title(f'Percentage of {title}')
    ax[1].set_xlabel(title)
    ax[1].set_ylabel('Percentage')
    ax[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Plot for each categorical column
for col in categorical_columns:
    plot_categorical_data(col, col)
```

```python
# Function to plot histograms and box plots for numeric columns
def plot_numeric_data(column, title):
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))

    # Histogram
    sns.histplot(df[column], kde=True, ax=ax[0], color='blue')
    ax[0].set_title(f'Histogram of {title}')
    ax[0].set_xlabel(title)
    ax[0].set_ylabel('Frequency')

    # Box Plot
    sns.boxplot(x=df[column], ax=ax[1], color='lightgreen')
    ax[1].set_title(f'Box Plot of {title}')
    ax[1].set_xlabel(title)

    plt.tight_layout()
    plt.show()

# Plot for each numeric column
for col in numeric_columns:
    plot_numeric_data(col, col)
```

```python
def combined_plot(column, title):
    freq_data = compute_frequencies_percentages(df[column])
    fig, axs = plt.subplots(2, 2, figsize=(16, 12))

    # Frequency Plot
    sns.barplot(x=freq_data.index, y=freq_data['Frequency'], ax=axs[0, 0], palette='viridis')
    axs[0, 0].set_title(f'Frequency of {title}')
    axs[0, 0].set_xlabel(title)
    axs[0, 0].set_ylabel('Frequency')
    axs[0, 0].tick_params(axis='x', rotation=45)

    # Percentage Plot
    sns.barplot(x=freq_data.index, y=freq_data['Percentage'], ax=axs[0, 1], palette='viridis')
    axs[0, 1].set_title(f'Percentage of {title}')
    axs[0, 1].set_xlabel(title)
    axs[0, 1].set_ylabel('Percentage')
    axs[0, 1].tick_params(axis='x', rotation=45)

    # Histogram
    sns.histplot(df[column], kde=True, ax=axs[1, 0], color='blue')
    axs[1, 0].set_title(f'Histogram of {title}')
    axs[1, 0].set_xlabel(title)
    axs[1, 0].set_ylabel('Frequency')

    # Box Plot
    sns.boxplot(x=df[column], ax=axs[1, 1], color='lightgreen')
    axs[1, 1].set_title(f'Box Plot of {title}')
    axs[1, 1].set_xlabel(title)

    plt.tight_layout()
    plt.show()

# For categorical columns (assuming they are numeric or you convert them)
for col in categorical_columns:
    combined_plot(col, col)

# For numeric columns
for col in numeric_columns:
    plot_numeric_data(col, col)
```
